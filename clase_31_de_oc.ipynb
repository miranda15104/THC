```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x =np.linspace(-2*np.pi, 6*np.pi, 100)
seno=np.sin(x)



plt.subplot(1, 2, 1) # primero columnas, segundo fila
plt.plot(x,seno,"*:m",ms=5)
plt.title("Seno")
plt.grid()

#plot 2:
x = np.linspace(-2*np.pi, 6*np.pi, 100)
Coseno=np.cos(x)

plt.subplot(1, 2, 2)
plt.plot(x, Coseno ,"o:r", ms=5)
plt.title("coseno")
plt.grid()

plt.suptitle("funciones")

plt.show()
```
