```python
def primo(numero): #definimos nuestra funcion
    if numero <= 1:
        return "Error en el Número"

    for i in range(2, int(numero**0.5) + 1):
        if numero  % i  == 0:   #%2 para saber si es primo o no
            return False

    return True

# Solicitar entrada del usuario y ver los casos inválidos
while True:
    try:
        entrada = int(input("Ingrese un número entero mayor que 1: "))
        resultado = primo(entrada)

        if resultado == "Error en el Número":
            print(resultado)
        elif resultado:
            print("True")
        else:
            print("False")

        break  # Salir del bucle si la entrada es válida
    except ValueError:
        print("Entrada inválida. Por favor, ingrese un número entero.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Creamos un arreglo de números enteros entre 0 y 150
numerosEnteros = np.random.randint(0, 151, 100)

# Función para calcular el promedio
promedio = lambda nums: sum(nums) / len(nums)

# Función para calcular la varianza
varianza = lambda nums: sum((x - promedio(nums))**2 for x in nums) / len(nums)

# Función para calcular la desviación estándar
desviacionEstandar = lambda nums: varianza(nums)**0.5

# Calcular el promedio, la varianza y la desviación estándar
promedioResultado = promedio(numerosEnteros)
varianzaResultado = varianza(numerosEnteros)
DesviacionEstandarResultado = desviacionEstandar(numerosEnteros)

# Imprimimos resultados
print("Promedio:", promedioResultado)
print("Varianza:", varianzaResultado)
print("Desviación Estándar:", DesviacionEstandarResultado)

# Visualizar los datos en un histograma
plt.hist(numerosEnteros, edgecolor="pink", color="purple", alpha=0.9)
plt.title("Histograma de Números Enteros")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()

# Calcular la suma de números pares e impares
suma_pares = sum(x for x in numerosEnteros if x % 2 == 0)
suma_impares = sum(x for x in numerosEnteros if x % 2 != 0)

# Imprimir resultados
print("Suma de números pares:", suma_pares)
print("Suma de números impares:", suma_impares)

# Calcular múltiplos de 2, 3 y 5
multiplos_2 = [x for x in numerosEnteros if x % 2 == 0]
multiplos_3 = [x for x in numerosEnteros if x % 3 == 0]
multiplos_5 = [x for x in numerosEnteros if x % 5 == 0]

# Imprimir resultados
print("Múltiplos de 2:", multiplos_2)
print("Múltiplos de 3:", multiplos_3)
print("Múltiplos de 5:", multiplos_5)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y(x, C)
def y(x, C):
    return (x - 1) + C * np.exp(-x)

# Valores de x
x_values = np.linspace(-2, 4, 400)

# Graficar para C > 0, C = 0 y C < 0
plt.figure(figsize=(8, 6))

# Para C > 0 (por ejemplo, C = 1)
plt.plot(x_values, y(x_values, 1), label="C > 0", color="purple")

# Para C = 0
plt.plot(x_values, y(x_values, 0), label="C = 0", color="orange")

# Para C < 0 (por ejemplo, C = -1)
plt.plot(x_values, y(x_values, -1), label="C < 0", color="pink")

# Añadir detalles al gráfico
plt.title("Solución de la Ecuación Diferencial")
plt.xlabel("x")
plt.ylabel("y")

plt.grid(color = "gray", linestyle = "--", linewidth = 0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Graficamos la función con línea continua para un rango de -10 a 10 con más de 1000 muestras
x_values = np.linspace(-10, 10, 1000)
y_values = (3 * x_values + 7) / ((x_values**2) - 4)

#Graficamos las asíntotas horizontales y verticales
plt.axhline(0, color="purple",linestyle="--", label="Asíntota Horizontal")
plt.axvline(0, color="pink", linestyle="-", label="Asíntota Vertical")

#Establecemos límites en el eje "y" para evitar valores muy altos o bajos cerca de las asíntotas verticales
plt.ylim(-10, 10)

#Graficamos la función con línea continua y añadimos la leyenda
plt.plot(x_values, y_values, label="3x + 7/x^2 - 4", color="green")
plt.legend()

#Detalles de nuestro gráfico
plt.title("Gráfica de la Función con Asíntotas")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
